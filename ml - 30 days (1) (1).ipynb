```python
# Scatterplot for Age vs Fare
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', data=df, hue='Survived', palette='Set1')
plt.title('Scatterplot of Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived')
plt.show()

# Correlation coefficient
correlation = df[['Age', 'Fare']].corr()
print("Correlation between Age and Fare:\n", correlation)
```

```python
# Boxplot for Age vs Survived
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='Survived', y='Age', data=df, palette='Set2')
plt.title('Age vs Survived')
plt.xlabel('Survived')
plt.ylabel('Age')

# Boxplot for Fare vs Survived
plt.subplot(1, 2, 2)
sns.boxplot(x='Survived', y='Fare', data=df, palette='Set2')
plt.title('Fare vs Survived')
plt.xlabel('Survived')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

```python
#categorical-categorical(bar plot, heatmap)
```

```python

# Bar plot for Pclass vs Survived
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='Set2')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

```python

# Create a cross-tabulation of Pclass and Survived
cross_tab = pd.crosstab(df['Pclass'], df['Survived'])
```

```python
# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cross_tab, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Pclass vs Survived')
plt.xlabel('Survived')
plt.ylabel('Passenger Class (Pclass)')
plt.show()
```
